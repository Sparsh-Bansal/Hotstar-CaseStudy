```python
# Importing the packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Reading The data
media=pd.read_csv('mediacompany.csv')
media.head()   # Print the first 5 rows
```

```python
# Dealing with the missing data
media=media.drop(['Unnamed: 7'],axis=1)
media.head()
```

```python
# Analysing the data
media.describe()
```

```python
# Getting imformation about the data
media.info()
```

```python
media.head()
```

```python
# Converting the date format to YYYY-MM-DD format
media['Date']=pd.to_datetime(media['Date'])
media.head()
```

```python
# Adding another column
media['day']= [i+1 for i in range(media.shape[0])]
```

```python
media.head()
```

```python
media.plot.line(x='day',y='Views_show')
```

```python
# Scatter plot to check the relationship between No. of days spent vs no. of views on a particular show
plt.scatter(media.day,media.Views_show,marker='d',c='red',alpha=0.75,s=np.pi*9)
plt.title('Scatter plot-Days vs Views Show',fontsize=20)
plt.xlabel('No.of Days',fontsize=15)
plt.ylabel('Views on a particular show',fontsize=15)
plt.show()
```

```python
# plot for days vs Views_show and days vs Ad_impressions

fig = plt.figure()
host = fig.add_subplot(1,1,1)

par1 = host.twinx()
par2 = host.twinx()

host.set_xlabel("Day")
host.set_ylabel("View_Show")
par1.set_ylabel("Ad_impression")

color1 = plt.cm.viridis(0.1)
color2 = plt.cm.viridis(0.9)
color3 = plt.cm.viridis(0)

p1, = host.plot(media.day,media.Views_show, color=color1,label="View_Show")
p2, = par1.plot(media.day,media.Ad_impression,color=color2, label="Ad_impression")

lns = [p1, p2]
host.legend(handles=lns, loc='best')

# right, left, top, bottom
par2.spines['right'].set_position(('outward', 60))      
# no x-ticks                 
par2.xaxis.set_ticks([])
# Sometimes handy, same for xaxis
#par2.yaxis.set_ticks_position('right')

host.yaxis.label.set_color(p1.get_color())
par1.yaxis.label.set_color(p2.get_color())

plt.savefig("pyplot_multiple_y-axis.png", bbox_inches='tight')
```

```python
X=media[['Ad_impression_2','weekend','Character_A']]
Y=media['Views_show']
import statsmodels.api as sm
X=sm.add_constant(X)
pred_views=lm6.predict(X)
```

```python
c=[i for i in range(1,81)]
fig=plt.figure()
plt.plot(c,Y,color='blue',linewidth=1.5,linestyle='-')
plt.plot(c,pred_views,color='red',linewidth=1.5,linestyle='-')
fig.suptitle('Actual vs Predicted',fontsize=20)
plt.xlabel('Index',fontsize=18)
plt.ylabel('Views',fontsize=16)
plt.show()
```

```python
c=[i for i in range(1,81)]
fig=plt.figure()
plt.plot(c,Y-pred_views,color='blue',linewidth=3.5,linestyle='-')
fig.suptitle('Error Terms',fontsize=20)
plt.xlabel('Index',fontsize=18)
plt.ylabel('Actual_views-predictes_views',fontsize=16)
plt.show()
```
